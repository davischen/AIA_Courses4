```python
# 大多資料集中在網管拆開分散到大類
def network_b_to_a(row):
    if '網管' in row['category_a']:
        return row['category_b']
    else:
        return row['category_a']

df_category_a['network_b_to_a'] = df_category_a.apply(network_b_to_a, axis=1)
df_category_a[df_category_a['category_a'] == '網管'].head()
```

```python
# 降低特別大的類別 -- 因為資料還是集中在c類上 所以由c類下去隨機抽樣
print('電腦網路不通')
print(df_category_a.shape)
df_sample = df_category_a[df_category_a['category_c'] == '電腦網路不通'].sample(n=6000, random_state=1)
df_category_a = df_category_a[df_category_a['category_c'] != '電腦網路不通']
frames = [df_category_a, df_sample]
df_category_a = pd.concat(frames)
print(df_category_a.shape)
# print('Outlook')
# print(df_category_a.shape)
# df_sample = df_category_a[df_category_a['category_c'] == 'Outlook'].sample(n=4000, random_state=1)
# df_category_a = df_category_a[df_category_a['category_c'] != 'Outlook']
# frames = [df_category_a, df_sample]
# df_category_a = pd.concat(frames)
# print(df_category_a.shape)
print('作業系統問題')
print(df_category_a.shape)
df_sample = df_category_a[df_category_a['category_c'] == '作業系統問題'].sample(n=800, random_state=1)
df_category_a = df_category_a[df_category_a['category_c'] != '作業系統問題']
frames = [df_category_a, df_sample]
df_category_a = pd.concat(frames)
print(df_category_a.shape)
```

```python
# 資料量較少與易混淆類別 合併
# 其他類要屏除 主要大類中會出現的data
def combie_category_a(row):
    if row['network_b_to_a'] in ['SAP-168(QCI)','SAP-2X8(QSMC)']:
        return 'SAP'
    elif row['network_b_to_a'] in ['檔案共用或FTP','視訊會議設備','作業系統與印表機','電腦防毒','商務用Skype']:
        return '其他'
    elif row['network_b_to_a'] in ['BPM平台系統','R&D資訊系統','內部資訊系統','CAMP']:
        return '資訊系統'
    elif row['network_b_to_a'] in ['VPN連線','網路']:
        return '網路'
    else:
        return row['network_b_to_a']

df_category_a['category'] = df_category_a.apply(combie_category_a, axis=1)
df_category_a.head()
```

```python
df_category_a = df_category_a.drop(['network_b_to_a'], axis = 1)
# df_category_a = df_category_a.drop(['category_a','category_b','category_c','category_d','network_b_to_a'], axis = 1)
```
