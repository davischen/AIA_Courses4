```python
"""
把字串全形轉半形
"""
def full_to_half(s):
    n = []
#     s = s.decode('utf-8')
    for char in s:
        num = ord(char)
        if num == 0x3000:
            num = 32
        elif 0xFF01 <= num <= 0xFF5E:
            num -= 0xfee0
        num = chr(num)
        n.append(num)
    return ''.join(n)

df_all[clean_col_name] = df_all[clean_col_name].apply(full_to_half)
```

```python
"""
清理員工中英文姓名
"""
# 讀取員工資料
df_employee = pd.read_csv('data/employee.csv')
employee_cht_name_list = list(df_employee['employee_cht_name'])
employee_eng_name_list = list(df_employee['employee_eng_name'])
employee_id_name_list = list(df_employee['employee_id'])
print(employee_cht_name_list[:3])
print(employee_eng_name_list[:3])
print(employee_id_name_list[:3])
print(f"df_employee shape : {df_employee.shape}")
df_employee = df_employee[['camp_user_id','employee_cht_name','employee_eng_name','employee_id']].head(1)
df_employee.head(1)
```

```python
"""
清理 黃先生.陳小姐等字
"""
first_name = []
for name in employee_cht_name_list:
    if name[:1] in first_name:
        continue
    
    # 中文欄位放英文拿掉, 判斷是否為中文, 不是就跳過
    if name[:1].encode('UTF-8').isalpha():
        continue
        
    first_name.append(name[:1])
    
def regex_convertor_with_list_add_name(format_type, str_list):
    """
    把性別加上先生小姐再轉成正規表示法
    """
    if len(str_list) == 0:
        return None

    rule = ''
    if format_type == 'or':
        for word in str_list:
            if word == '':
                continue
            if type(word) != str :
                continue                

            rule = rule + word.lower() + '先生' + '|' + word.lower() + '小姐' + '|'
        result = r'(' + rule[0:-1] + ')'
    else:
        result = ''

    return result

cht_first_name_combie_regex = regex_convertor_with_list_add_name('or', first_name)
```

```python
def regex_convertor_with_list(format_type, str_list):
    """
    list文字轉成正規表示法
    """

    if len(str_list) == 0:
        return None

    rule = ''
    if format_type == 'or':
        for word in str_list:
            if word == '':
                continue
            if type(word) != str :
                continue                

            rule = rule + word.lower() + '|'
        result = r'(' + rule[0:-1] + ')'
    else:
        result = ''

    return result
    
def clean_employee(item):
    
    pattern = re.compile(cht_name_regex)
    item = re.sub(pattern, "", item)
    
    pattern = re.compile(eng_name_regex)
    item = re.sub(pattern, "", item)
    
    pattern = re.compile(emp_id_regex)
    item = re.sub(pattern, "", item)
    
    pattern = re.compile(cht_first_name_combie_regex)
    item = re.sub(pattern, "", item)
        
    return item

cht_name_regex = regex_convertor_with_list('or', employee_cht_name_list)
eng_name_regex = regex_convertor_with_list('or', employee_eng_name_list)
emp_id_regex = regex_convertor_with_list('or', employee_id_name_list)
df_all[clean_col_name] = df_all[clean_col_name].apply(clean_employee)
```
